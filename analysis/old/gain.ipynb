```python
import numpy as np
import matplotlib.pyplot as plt

input_amp=0.49
blr=[0.1196,0.1183,0.1236,0.1181] # h, m, l, set
thresh=[[0.439,0.434,0.431],
        [0.431,0.425,0.422],
        [0.427,0.421,0.418],
        [0.428,0.422,0.418],
        [0.448,0.441,0.439],
        [0.437,0.431,0.428],
        [0.437,0.431,0.428],
        [0.437,0.431,0.428],
        [0.465,0.459,0.456],
        [0.452,0.446,0.441],
        [0.453,0.447,0.443],
        [0.453,0.446,0.443],
        [0.437,0.431,0.428],
        [0.428,0.422,0.419],
        [0.428,0.422,0.418],
        [0.428,0.422,0.418],
        [0.431,0.425,0.422],
        [0.422,0.415,0.412],
        [0.421,0.414,0.411],
        [0.418,0.412,0.409]]

amp=[]
gain=[]
for i in range(len(thresh)):
    avg=0
    for j in range(len(thresh[i])):
        avg+=thresh[i][j]-blr[j]
    avg=avg/len(thresh[i])
    amp.append(avg)
    gain.append(avg/input_amp)

channels = ['CH1','CH2','CH3','CH4','CH5','CH6','CH7','CH8','CH9','CH10','CH11','CH12','CH13','CH14','CH15','CH16','CH17','CH18','CH19','CH20']
```

```python

# Sample data
std_dev=np.std(gain)
# Find the index of the bar with the lowest value
min_index = gain.index(min(gain))
max_index = gain.index(max(gain))
# Create bar plot
bars = plt.bar(channels,gain)
# Set the color of the bar with the lowest value to gold
bars[min_index].set_color('tomato')
bars[max_index].set_color('lightgreen')
# Add value label on top of the bar with the lowest value, rotated 90 degrees
plt.text(min_index, gain[min_index]+0.03, f"{gain[min_index]:.2f}", ha='center', rotation=90)
plt.text(max_index, gain[max_index]+0.03, f"{gain[max_index]:.2f}", ha='center', rotation=90)
# Rotate the x-axis labels
plt.xticks(rotation=90)
plt.ylabel('Gain')
plt.title('CASB "Unity" Path Channel Gains')
plt.ylim(0,1)
# Add standard deviation to the legend
plt.legend([f'$\\sigma= {std_dev:.2f}$'])
# Show the plot
plt.show()
```
